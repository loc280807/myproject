```python
#bai6
line = input()
a,b,c = map(float, line.split())
p = (a+b+c)/2
if (p-a >0) and (p-b>0) and (p-c)>0:
    S =(p * (p - a) * (p - b) * (p - c))**(1/2)
    print(S)
else:
    print("ko ton tai tam giac")
```
