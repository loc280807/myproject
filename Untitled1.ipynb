```python

def dem(i):
    dem = 0
    while i != 1:
        if i % 2 == 0:
            i = i // 2
        else:
            i = i * 3 + 1
        dem += 1
    return dem

if __name__ == "__main__":
    n = int(input())
    nhap = 0
    for i in range(1, n + 1):
        if dem(i) > nhap:
            nhap = dem(i)
    print(nhap)
```
