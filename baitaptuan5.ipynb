```python
#Viết hàm kiểm tra một số có phải là số nguyên tố hay không? 
import math

def ktrasnt(n):
    if n < 2:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

print(ktrasnt(17))  # True
```

```python
#VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên

def shh(n):
    sum = 0
    for i in range(1,  n+ 1):
        if n%i == 0:
            sum = sum + i
            if sum == n:
                return True

shh(6)
```

```python
def pheptoan(n1, n2):
    pheptru = n1 - n2 
    phepcong = n1 + n2
    phepchia = n1 / n2
    phepnhan = n1 * n2
    return round(pheptru, 2), round(phepcong, 2), round(phepchia, 2), round(phepnhan, 2)

pheptoan(7.36, 5)
```

```python
def nhiphancua(n):
    A = []
    while n > 0:
        du = n % 2
        A.append(du)
        n = n // 2
    A.reverse()     
    return A

def chuanhoa(a,b):
    A = nhiphancua(a)
    B = nhiphancua(b)
    while len(A) != len(B): 
        if len(A) < len(B): 
            A= [0]+ A
        else:
            B=[0]+B
    return A,B

chuanhoa(13,5)

def hamming(a,b):
    A, B = chuanhoa(a, b)
    sum = 0
    for i in range(len(A)):
        if A[i] != B[i]:
            sum +=1
    return sum

hamming(48226825258,5262548488)
```

```python
def chuanhoakitu(a):
    return [ord(c) for c in a]

def dang_cau(a, b):
    # 1. Kiểm tra độ dài
    if len(a) != len(b):
        return False
    
    # Trường hợp chuỗi rỗng
    if not a: 
        return True

    A = chuanhoakitu(a)
    B = chuanhoakitu(b)
    
    # 2. Tính độ lệch k dựa trên cặp ký tự đầu tiên
    k = B[0] - A[0]
    
    # 3. Duyệt qua từng phần tử để kiểm tra
    for i in range(len(A)):
        # Nếu có bất kỳ cặp nào không khớp độ lệch k -> Sai
        if A[i] + k != B[i]:
            return False
            
    # Nếu chạy hết vòng lặp mà không có lỗi -> Đúng
    return True

# Test
# 'a'(97) -> 'z'(122): k = 25
# 'b'(98) -> 'x'(120): k = 22 => Sai logic (k không đổi)
print(dang_cau('abca', 'zbxz')) # Output: False

# Test trường hợp đúng (dịch chuyển 1 đơn vị)
print(dang_cau('abca', 'bcd'))   # Output: True
```
