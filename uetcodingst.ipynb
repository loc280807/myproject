```python
# Viết hàm find_duplicates(lst) trả về dictionary với key là phần từ trùng và value là số lần xuất hiện (chỉ những phần tử xuất hiện > 1 lần)

def find_duplicates(lst):
    dic = {}
    for i in lst:
        if i in dic:
            dic[i] += 1
        else:
            dic[i] = 1

    return {k: v for k, v in dic.items() if v > 1}

lst = list(map(int, input().split()))
print(find_duplicates(lst))
```

```python
# Viết chương trình yêu cầu người nhập vào một chuỗi ngoặc. Kiểm tra chuỗi có hợp lệ không 
# ()[]{} Valid
# (] Invalid
# ([)] Invalid

def kiem_tra_ngoac(chuoi):
    stack = []
    chuoi_ngoac ={')': '(', ']': '[', '}': '{'}

    for ky_tu in chuoi:
        if ky_tu in chuoi_ngoac:
            top_element = stack.pop() if stack else '#'
        else:
            stack.append(ky_tu)
    
    return not stack

if __name__ == "__main__":
    # Nhập dữ liệu từ bàn phím
    s = input("Nhập chuỗi ngoặc cần kiểm tra: ")
    
    if kiem_tra_ngoac(s):
        print(f"Chuỗi '{s}' -> Hợp lệ (Valid)")
    else:
        print(f"Chuỗi '{s}' -> Không hợp lệ (Invalid)")


```

```python
def sum_odds(a):
  if not a:
    return 0

  firt = a[0]
  firt__ = a[1:]

  if firt%2 != 0:
    return firt + sum_odds(firt__)
  else:
    return sum_odds(firt__)
      

n = int(input().strip())
a = list(map(int, input().split())) if n > 0 else []
print(sum_odds(a))
```
