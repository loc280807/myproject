```python
# Nhâ p vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma ̣
# trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải

m  = int(input("Nhập số hàng:"))
n  = int(input("Nhập số cột:"))
matrix = []

for i in range(n):
    hangmoi = [] 
    for j in range(m):
        giatri = int(input(f"Nhập vào giá trị {[i],[j]}:"))
        hangmoi.append(giatri)
    matrix.append(hangmoi)

print("Ma trận đã nhập:")
for i in matrix:
    for j in i:
        print(f'{j:>4}' ,end='  ')
    print()
```

```python
# Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra đường chéo chính và
# đường chéo phụ của mảng

n  = int(input("Nhập số hàng :"))
matrix = []

for i in range(n):
    hangmoi = [] 
    for j in range(n):
        giatri = int(input(f"Nhập vào giá trị {[i],[j]}:"))
        hangmoi.append(giatri)
    matrix.append(hangmoi)

print("Ma trận đã nhập:")
for i in matrix:
    for j in i:
        print(f'{j:>4}' ,end='  ')
    print()

for i in range(n):
    for j in range(n):
        if i != n-1-j and i != j:
            matrix[i][j] = ''

print("duong cheo chinh va phu:")

for i in matrix:
    for j in i:
        print(f'{j:>4}' ,end='  ')
    print()
```

```python
# Nhâ ̣p vào mảng n hàng, m cột số nguyên và số nguyên k (n, m, k nhập trên ̣
# cùng một dòng, cách nhau dấu cách). Tính và in ra tổng các phần tử trên cột
# k (hàng k) 

n, m, k = map(int, input().split())
matrix = []
for i in range(n):
    hangmoi = [] 
    for j in range(n):
        giatri = int(input(f"Nhập vào giá trị {[i],[j]}:"))
        hangmoi.append(giatri)
    matrix.append(hangmoi)

print("Ma trận đã nhập:")
for i in matrix:
    for j in i:
        print(f'{j:>4}' ,end='  ')
    print()

dem = 0
for i in range(len(matrix)):
    dem = dem + matrix[i][k-1]
print(dem)
```

```python
# Viết hàm đưa vào 1 list số nguyên, tìm và trả về vị trí có giá trị lớn nhất
# trong list
A = list(map(int, input().split()))
so_lon_nhat=A[0]
for i in A:
    if i > A[0]:
        so_lon_nhat = i
print(so_lon_nhat)
```

```python
# Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và
# trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về
# -1

def kiemtra(A,k):
    for i in range(len(A)):
        if k == A[i]:
            return i
    return -1

A = list(map(int,input().split()))
k= int(input())
print(kiemtra(A,k))
```
