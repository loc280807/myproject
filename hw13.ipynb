```python
# w11a1
def reverse(path: str) -> int:
    with open(path, 'r') as f:
        content = f.read().strip()

    try:
        content_rever = content[::-1]
        return int(content_rever)
    except: 
        return "Nhap lai file"                     

# --- Chạy thử ---
ket_qua = reverse('w11a1.txt')
print(ket_qua)
```

```python
#w11a10
def getMoney(path:str) -> list[int]:
    with open(path, 'r') as f:
        content = f.read().split()
    A = [int(x) for x in content]
    return A

def rob(hourses: list[int]):
    n = len(hourses)
    if n == 0:
        return 0
    if n == 1:
        return hourses[0]
    if n == 2:
        return max(hourses[0], hourses(1))
    dp = [0]*n
    dp[0] = hourses[0]
    dp[1] = max(hourses[0], hourses[1])

    for i in range (2,n):
        dp[i] = max(hourses[i] + dp[i-2], dp[i-1])
    return dp[-1]

print(getMoney("w11a10.txt"))
hourses = getMoney("w11a10.txt")

rob(hourses)
```

```python
def fibonacci_word(n):
    fib = ["a", "b"]
    for i in range(2, n+1):
        fib.append(fib[i-1]+fib[i-2])
    with open("output.txt", "w") as f:
        for i in range(n+1):
            f.write(f"f({i}) = {fib[i]} \n")

fibonacci_word(10)
```

```python
def count(path, character):
    with open(path, "r") as f:
        doc = f.read().strip()
        cnt = 0
        for i in doc:
            if i == character:
                cnt = cnt +1 
        return cnt
    
count("w11a5.txt", "k")
```

```python
def productExceptSelf(path: str) -> list[int]:

    with open(path, 'r') as f:
        content = f.read().split()
        nums = [int(x) for x in content]


    n = len(nums)
    if n <= 1:
        return nums
    result = [1] * n

    left_product = 1
    for i in range(n):
        result[i] = left_product
        left_product *= nums[i]

    right_product = 1
    for i in range(n - 1, -1, -1):
        result[i] *= right_product
        right_product *= nums[i]

    return result

productExceptSelf("w11a17.txt")
```

```python
def ki_tu(chuoi):
    danh_sach = {}
    for i in chuoi:
        if not i.isalpha():
            continue
        ky_tu_chuan = i.lower()
        if ky_tu_chuan in danh_sach:
            danh_sach[ky_tu_chuan] += 1
        else:
            danh_sach[ky_tu_chuan] = 1 
    return danh_sach

def areAnagrams(path1: str, path2: str):
    try:
        with open(path1, "r", encoding="utf-8") as f1:
            c1 = f1.read().strip()
        with open(path2, "r", encoding="utf-8") as f2:
            c2 = f2.read().strip()
        
        # So sánh 2 bảng tần suất
        if ki_tu(c1) == ki_tu(c2):
            return True
        return False
    except FileNotFoundError:
        return False

print(areAnagrams("w11a4_1.txt", "w11a4_2.txt"))
```

```python
def countCharacter(path: str, letter: str):
    with open(path, "r") as f:
        chuoi = f.read().strip()

    danh_sach = {}
    for i in chuoi:
        if not i.isalpha():
            continue
        ky_tu_chuan = i.lower()
        if ky_tu_chuan in danh_sach:
            danh_sach[ky_tu_chuan] += 1
        else:
            danh_sach[ky_tu_chuan] = 1 
    
    if letter in danh_sach:
        return danh_sach[letter]
    return 0
    
countCharacter("w11a6.txt", 'e')

```
