```python
while True:
    try:
        n = int(input("Nhập số nguyên dương: "))
        
        if n <= 0:
            raise ValueError("Số phải lớn hơn 0")   # tự ném lỗi để bắt ở except
        
        break   # hợp lệ thì thoát khỏi vòng lặp

    except ValueError as e:
        print("Lỗi:", e)
        print("Vui lòng nhập lại!\n")

print("Bạn đã nhập đúng:", n)
```

```python
def fib(n):
    if n <= 1:
        return 1
    return fib(n - 1) + fib(n - 2)

def find_fib(A, n=1):
    if fib(n) > A:
        return fib(n-1)
    return find_fib(A, n+1)

A = int(input("Nhập A: "))
print(find_fib(A))


```
