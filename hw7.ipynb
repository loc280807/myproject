```python
def binary_search(arr,target):
    left,right = 0,len(arr)-1
    while left <= right:
        mid = (left+right)//2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid +1 
        else:
            right = mid -1 
    return -1

arr = list(map(int, input().split()))
target = int(input())
print(binary_search(arr, target))
```

```python
def count_occurrences(arr, x):
    dem = 0
    for i in range(len(arr)):
        if arr[i] == x:
            dem +=1
    return dem


arr = list(map(int, input().split()))
x = int(input())
count_occurrences(arr,x)
```

```python
def w7a5(arr, target):
    arr.sort()
    count = 0
    left, right = 0, len(arr) - 1
    while left < right:
        total = arr[left] + arr[right]
        if total < target:
            left += 1
        elif total > target:
            right -= 1
        else:
            if arr[left] == arr[right]:
                n = right - left + 1
                count += (n * (n - 1)) // 2
                break # Đã đếm hết, thoát vòng lặp
            else:
                # Đếm số lượng số giống arr[left]
                l_count = 1
                while left + 1 < right and arr[left] == arr[left+1]:
                    l_count += 1
                    left += 1
                
                # Đếm số lượng số giống arr[right]
                r_count = 1
                while right - 1 > left and arr[right] == arr[right-1]:
                    r_count += 1
                    right -= 1
                
                # Cộng số tổ hợp: ví dụ có 3 số 2 và 4 số 3 -> 3*4 = 12 cặp
                count += (l_count * r_count)
                
                # Di chuyển tiếp
                left += 1
                right -= 1
    print(count)

if __name__ == "__main__":
    arr = list(map(int, input().split()))
    target = int(input())
    w7a5(arr,target)
```
