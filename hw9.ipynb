```python
class Cylinder:
    pi=3.14
    
    def __init__(self,r,h):
        self.x=r
        self.y=h
    def S(self):
        Sbm=2*Cylinder.pi*self.x*self.y+Cylinder.pi*self.x**2
        return f"{Sbm:.2f}"
    def V(self):
        V_=Cylinder.pi*(self.x)**2*self.y
        return f"{V_:.2f}"
    def __str__(self):
        return f"{self.S()} {self.V()}"
a,b=map(float,input().split())
hinhtru=Cylinder(a,b)
print(hinhtru)
```

```python
class date:
    def __init__(self,day,month,year):
        self.d=day
        self.m=month
        self.y=year
    def nam_nhuan(self):
        return self.y%400==0 or (self.y%4==0 and self.y%100!=0)
    def ngay_cuoi_trong_thang(self):
        c=[1,3,5,7,8,10,12]
        d=[4,6,9,11]
        if self.m in c:
            return 31
        elif self.m in d:
            return 30
        elif self.m==2:
            if self.nam_nhuan():
                return 29
            else:
                return 28
    def ngay_tiep_theo(self):
        if self.d == self.ngay_cuoi_trong_thang():
            self.d=1
            self.m+=1
            if self.m>12:
                self.m=1
                self.y+=1
        else:
            self.d+=1
    def __str__(self):
        return f"{self.d:02d} / {self.m:02d} / {self.y}"
d,m,y=map(int,input().split())
ngay_thang_nam=date(d,m,y)
print("ngay hien tai: ",ngay_thang_nam)
ngay_thang_nam.ngay_tiep_theo()
print("ngay tiep theo : ",ngay_thang_nam)
```

```python
import math
class truc_toa_do:
    def __init__(self,a,b):
        self.x=a
        self.y=b
    def goc_toa_do(self):
        return self.x==0 and self.y==0
    def nam_tren_truc_tung(self):
        return self.x==0 and self.y!=0
    def nam_tren_truc_hoanh(self):
        return self.y==0 and self.x!=0
    def khong_nam_tren_truc_nao(self):
        return self.y!=0 and self.x!=0
    def khoang_cach_den_o(self):
        return math.sqrt((self.x)**2+(self.y)**2)
    def __str__(self):
        if self.goc_toa_do():
            vi_tri="M là gốc tọa độ"
        elif self.nam_tren_truc_hoanh():
            vi_tri="M nằm trên trục hoành"
        elif self.nam_tren_truc_tung():
            vi_tri="M nằm trên trục tung"
        elif self.khong_nam_tren_truc_nao():
            vi_tri="M không nằm trên trục nào"
        return f"{vi_tri}. Khoảng cách đến gốc tạo độ là {self.khoang_cach_den_o():.2f}"
a,b=map(int,input().split())
M=truc_toa_do(a,b)
print(M)
```

```python
class Rectangle:
    def __init__(self,a,b):
        self.w=a
        self.h=b
    def area(self):
        return f"dien tich truoc khi phong to la: {self.w * self.h}"
    def perimeter(self):
        return f"chu vi truoc khi phong to la: {(self.w+self.h)*2}"
    def scale(self,k):
        self.w=self.w*k
        self.h=self.h*k
        S2=self.w*self.h
        C2=(self.w+self.h)*2
        return (f"dien tich sau khi phong to la: {S2:.2f} \n"
                f"chu vi sau khi phong to la: {C2:.2f}")
w,h,k=map(float,input().split())
c=Rectangle(w,h)
print(c.area())
print(c.perimeter())
print(c.scale(k))
```

```python
from fractions import Fraction
class Fractionn:
    def __init__(self,ts,ms):
        self.ts=ts
        self.ms=ms
        if self.ms==0:
            return "khong ton tai phan so nay"
    def __add__(self,ts2,ms2):
        return Fraction(self.ts,self.ms) + Fraction(ts2,ms2)
    def __sub__(self,ts2,ms2):
        return Fraction(self.ts,self.ms) - Fraction(ts2,ms2)
    def __mul__(self,ts2,ms2):
        return Fraction(self.ts,self.ms) * Fraction(ts2,ms2)
    def __div__(self,ts2,ms2):
        return Fraction(self.ts,self.ms) / Fraction(ts2,ms2)
a,b=map(int,input().split())
z=input()
c,d=map(int,input().split())
e=Fractionn(a,b)
if z=="+":
    print(e.__add__(c,d))
elif z=="-":
    print(e.__sub__(c,d))
elif z=="*":
    print(e.__mul__(c,d))
elif z=="/":
    print(e.__div__(c,d))
```

```python
class Calculator:
    def __init__(self,a,b):
        self.x=a
        self.y=b
    def add(self):
        return self.x + self.y
    def subtract(self):
        return self.x - self.y
    def product(self):
        return self.x * self.y
    def divide(self):
        if self.y==0:
            return "khong the chia cho 0"
        return self.x / self.y
    def power(self):
        return self.x**self.y
    def mod(self):
        return self.x % self.y
    def set_number(self,e,f):
        self.x=e
        self.y=f
        return self.x and self.y
    def dung_chuong_trinh(self):
        while self.x=="yes" or self.y=="yes":
            break
    def __str__(self):
        return (f"Phep cong: {self.add()}\n"
            f"Phep tru: {self.subtract()}\n"
            f"Phep nhan: {self.product()}\n"
            f"Phep chia: {self.divide()}\n")
a,b=map(float,input().split())
M=Calculator(a,b)
flag=False
while True:
    if not flag:
        print(M)
    n=input()
    if n.lower()=="yes":
        print("bạn đã chọn dừng chương trình")
        break
    doi_so=input()
    if doi_so=="ok":
        e,f=map(float,input().split())
        M.set_number(e,f)
        print("đổi 2 số ban đầu thành: ",e ,"và" ,f)
        flag=True
```

```python
class ShoppingCart:
    gio_hang=[]
    def __init__(self):
        pass
    def them_vao(self,name,price):
        kq={"name":name,"price":price}
        ShoppingCart.gio_hang.append(kq)
        print(f"da them vao {kq["name"]}")
    def xoa_1_sp(self,name):
        if len(ShoppingCart.gio_hang)==0:
            return "khoong co san pham de xoa"
        else:
            for i in ShoppingCart.gio_hang:
                if i["name"]== name:
                    ShoppingCart.gio_hang.remove(i["name"])
                    break
    def kiem_tra_rong(self):
        return len(ShoppingCart.gio_hang)==0
    def tinh_tong_tien(self):
        s=0
        for i in ShoppingCart.gio_hang:
            s+=int(i["price"])
        return f"tổng tiền là : {s}"
    def hien_thi_sp(self):
        for i in ShoppingCart.gio_hang:
            print(f"sản phẩm:{i['name']} có giá:{i['price']}")
    def xoa_gio_hang(self):
        ShoppingCart.gio_hang.clear()
        return "gio hang da duoc xoa"
cart=ShoppingCart()
while True:
    n=input()
    if n=="1":
        print("ch co gi trong gio hang")
    elif n=="2":
        t=input()
        g=int(input())
        cart.them_vao(t,g)
    elif n=="3":
        t=input()
        cart.xoa_1_sp(t)
    elif n=="4":
        if cart.kiem_tra_rong():
            print("rong")
        else:
            print("kh rong")
    elif n=="5":
        print(cart.tinh_tong_tien())
    elif n=="6":
        cart.hien_thi_sp()
    elif n=="7":
        print('da xoa het gio hang')
    dungct=input()
    if dungct.lower()=="yes":
        break
```

```python
class student:
    def __init__(self,subject):
        self.a=[]
        self.s=subject
    def them_diem(self,score):
        self.a.append(float(score))
    def diem_tb(self):
        return sum(self.a)/len(self.a)
    def rank(self):
        if self.diem_tb()>=8:
            return "Excellent"
        elif self.diem_tb()<8 and self.diem_tb()>=6.5:
            return "good"
        elif self.diem_tb()<6.5 and self.diem_tb()>=5:
            return "Average"
        elif self.diem_tb()<5:
            return "poor"
    def __str__(self):
        return f"{self.s} {self.diem_tb():.2f} {self.rank()}"
a=input()
z=student(a)
n=int(input())
for i in range(n):
    subject,score=input().split()
    z.them_diem(float(score))
print(z)
```

```python
class BankAccount:                                  
    def __init__(self,owner,blance):               
        self.ten=owner                             
        self.sodu=float(blance)
        self.a=[]
    def deposit(self,n):
        self.x=float(n)
        self.sodu+=self.x
        return float(self.sodu)
    def withdraw(self,e):
        if self.sodu>=float(e):
            self.sodu-=float(e)
        return self.sodu
ten,sodu=input().split()
z=BankAccount(ten,sodu)
n=int(input())
for i in range(n):
    e=input().split()
    if e[0].lower()=="deposit":
        z.deposit(e[1])
    elif e[0].lower()=="withdraw":
        z.withdraw(e[1])
print(f"{z.sodu:.2f}")
```
