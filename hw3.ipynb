```python
h = abs(int(input()))
for i in range(h+1):
    l = h - i - 1 
    r = h + i - 1 
    for j in range(2*h-1):
        if j == l or j == r or i == h -1:
            print("*", end = " ")
        else:
            print(" ", end = " ")
    print()
```

```python
def tonguoc(n):
    tong = 0
    for i in range (1,n):
        if n % i == 0:
            tong = tong + i
    return tong

n = int(input())
if tong == n:
    print(n, "la so hoan hao")
else:
    print(n, "ko la so hoan hao")
```
