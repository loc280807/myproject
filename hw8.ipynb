```python
# Cho một mảng số nguyên không âm. Gọi 
# mảng đó là "mảng đẹp" nếu mỗi phần tử 
# x trong mảng xuất hiện đúng x lần. 
# Kiểm tra xem mảng có phải là mảng đẹp 
# hay không. Nếu không hãy xóa đi y số (y 
# là nhỏ nhất có thể) để mảng còn lại thành 
# mảng đẹp. 

def is_beautiful_array(arr):
    from collections import Counter
    
    count = Counter(arr)
    freq = Counter(count.values())

    for x in count:
        if count[x] != x:
            break
    else:
        return "YES" 
    to_delete = 0
    for x in count:
        if count[x] > x:
            to_delete += count[x] - x
        elif count[x] < x:
            to_delete += count[x]
    
    return f"NO {to_delete}"

arr = list(map(int, input().split()))
print(is_beautiful_array(arr))
```

```python
# Tìm phần tử xuất hiện nhiều hơn n/2 lần
# trong mảng (nếu có) với độ phức tạp thời
# gian là O(n), độ phức tạp bộ nhớ là O(1),
# - Boyer–Moore. Nếu không có phần tử
# nào thỏa mãn thì không in gif

def majority_element(arr):
    candidate = None
    count = 0

    for num in arr:
        if count == 0:
            candidate = num
        count += (1 if num == candidate else -1)

    if arr.count(candidate) > len(arr) // 2:
        return candidate
    else:
        return "No majority element"    
    
arr = list(map(int, input().split()))
print(majority_element(arr))
```

```python
# Sắp xếp có điều kiện: Số chẵn đứng
# trước, số lẻ đứng sau, chẵn tăng dần, lẻ
# giảm dần
def conditional_sort(arr):
    evens = sorted([x for x in arr if x % 2 == 0])
    odds = sorted([x for x in arr if x % 2 != 0], reverse=True)
    return evens + odds 
arr = list(map(int, input().split()))
print(conditional_sort(arr))
```

```python
# Merge 2 mảng đã sắp xếp (có thể tăng
# dần, có thể giảm dần) mà không dùng
# bất cứ thuật toán sắp xếp nào thành một
# mảng cũng có cùng thứ tự sắp xếp như 2
# mảng ban đầu

def merge_sorted_arrays(arr1, arr2):
    if len(arr1) < 2:
        return arr1 + arr2
    ascending = arr1[0] < arr1[1]
    
    merged = []
    i, j = 0, 0
    while i < len(arr1) and j < len(arr2):
        if (arr1[i] <= arr2[j] and ascending) or (arr1[i] >= arr2[j] and not ascending):
            merged.append(arr1[i])
            i += 1
        else:
            merged.append(arr2[j])
            j += 1
    merged.extend(arr1[i:])
    merged.extend(arr2[j:])
    return merged

arr1 = list(map(int, input().split()))
arr2 = list(map(int, input().split()))
print(merge_sorted_arrays(arr1, arr2))
```

```python
# Cho dãy N số nguyên dương, tính tổng
# các số có k chữ số. Dòng đầu ghi N số
# nguyên dương, dòng sau ghi số nguyên
# dương k

def sum_numbers_with_k_digits(numbers, k):
    total = 0
    for num in numbers:
        if len(str(num)) == k:
            total += num
    return total

numbers = list(map(int, input().split()))
k = int(input())
print(sum_numbers_with_k_digits(numbers, k))
```

```python
# Cho N là ngày, T là tháng năm 2020,
# Tính xem đó là ngày thứ bao nhiêu của
# năm đó nếu đó là ngảy hợp lệ

def day_of_year(N, T):
    days_in_month = [31, 29, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    
    if T < 1 or T > 12 or N < 1 or N > days_in_month[T - 1]:
        return "Invalid date"
    
    day_count = sum(days_in_month[:T - 1]) + N
    return day_count
N, T = map(int, input().split())
print(day_of_year(N, T))
      
```
