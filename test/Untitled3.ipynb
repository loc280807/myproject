```python
#1
def unique_sorted_student_ids(list1, list2):
    # 1. Gộp 2 danh sách lại với nhau
    combined_list = list1 + list2
    
    # 2. Dùng set để loại bỏ phần tử trùng lặp
    unique_set = set(combined_list)
    
    # 3. Chuyển về list và sắp xếp tăng dần
    sorted_result = sorted(list(unique_set))
    
    return sorted_result

# Thay vì viết:
# l1 = map(str, input().split())

# Hãy viết:
l1 = list(map(str, input().split()))
l2 = list(map(str, input().split()))

# Gọi hàm
print(unique_sorted_student_ids(l1, l2))
```

```python
#3
def most_frequent_triple(triples):
    # 1. Kiểm tra nếu danh sách rỗng thì báo lỗi
    if not triples:
        raise ValueError("Input list cannot be empty")
    
    # 2. Tạo từ điển để đếm số lần xuất hiện
    counts = {}
    most_frequent = None
    max_count = 0
    
    for t in triples:
        if t in counts:
            counts[t] += 1
        else:
            counts[t] = 1
        
        # 3. Cập nhật phần tử xuất hiện nhiều nhất ngay trong vòng lặp
        if counts[t] > max_count:
            max_count = counts[t]
            most_frequent = t
            
    return most_frequent

triples = eval(input())
most_frequent_triple(triples)
```

```python
def reconcile_users(regitered, paid):
    l1 = set(paid) - set(regitered)
    l2 = set(regitered) - set(paid)
    l3 = set(i for i in regitered if i in paid)

    return l1, l2, l3

regitered = eval(input())
paid = eval(input())
reconcile_users(regitered, paid)
```

```python
def contains_duplicate(nums):
    danh_sach = {}
    for i in nums:
        if i in danh_sach:
            danh_sach[i] += 1
        else:
            danh_sach[i] = 1
            
    for key in danh_sach:
        if danh_sach[key] >= 2:
            return True
    return False
        
nums = eval(input())
contains_duplicate(nums)
```

```python
def squares_div_by3_not6(arr):
    danh_sach = [x**2 for x in arr if (x%3 == 0 and x%6 !=0)]
    return danh_sach

arr = eval(input())
squares_div_by3_not6(arr)
```

```python
def even_numbers_in_matrix(matrix):
    return [j for i in matrix for j in i if j % 2 == 0]
matrix = eval(input())
even_numbers_in_matrix(matrix)
```
