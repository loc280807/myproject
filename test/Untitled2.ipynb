```python
def w9a1(A,B):
    res = []
    i = j = 0

    while i < len(A) and j < len(B):
        if A[i] < B[j]:
            res.append(A[i])
            i += 1
        else:
            res.append(B[j])
            j +=1
    res.extend(A[i:])
    res.extend(B[j:])

    return res

A = list(map(int,input().split()))
B = list(map(int,input().split()))

w9a1(A,B)
```

```python
def merge_recursive(list1, list2):
    # 1. ĐIỂM DỪNG (Base Cases)
    # Nếu list1 hết, trả về phần còn lại của list2
    if not list1:
        return list2
    # Nếu list2 hết, trả về phần còn lại của list1
    if not list2:
        return list1

    # 2. BƯỚC ĐỆ QUY
    # So sánh phần tử đầu tiên (head)
    if list1[0] < list2[0]:
        # Nếu đầu list1 nhỏ hơn:
        # Lấy list1[0] + kết quả gộp của (phần còn lại list1 và list2)
        return [list1[0]] + merge_recursive(list1[1:], list2)
    else:
        # Ngược lại (đầu list2 nhỏ hơn hoặc bằng):
        # Lấy list2[0] + kết quả gộp của (list1 và phần còn lại list2)
        return [list2[0]] + merge_recursive(list1, list2[1:])

A = list(map(int,input().split()))
B = list(map(int,input().split()))
merge_recursive(A,B)
```

```python
import sys

def count_less_equal(matrix, target):
    """
    Đếm số phần tử trong ma trận nhỏ hơn hoặc bằng target.
    Độ phức tạp: O(N) nhờ tận dụng tính chất sắp xếp của ma trận.
    """
    n = len(matrix)
    m = len(matrix[0])
    count = 0
    
    # Bắt đầu từ góc dưới cùng bên trái
    row = n - 1
    col = 0
    
    while row >= 0 and col < m:
        if matrix[row][col] <= target:
            # Nếu phần tử hiện tại <= target
            # thì TẤT CẢ các phần tử phía trên nó (trong cùng cột) cũng <= target
            count += (row + 1)
            col += 1
        else:
            # Nếu phần tử hiện tại > target
            # ta cần tìm số nhỏ hơn, nên dịch lên dòng trên
            row -= 1
            
    return count

def find_kth_smallest_binary_search(matrix, k):
    n = len(matrix)
    if n == 0: return None
    m = len(matrix[0])
    
    low = matrix[0][0]
    high = matrix[n-1][m-1]
    
    while low < high:
        mid = (low + high) // 2
        
        # Đếm xem có bao nhiêu số <= mid
        count = count_less_equal(matrix, mid)
        
        if count < k:
            # Nếu số lượng chưa đủ k, nghĩa là mid quá nhỏ
            low = mid + 1
        else:
            # Nếu số lượng >= k, mid có thể là đáp án 
            # hoặc đáp án nằm bên trái
            high = mid
            
    return low

# --- TEST CASE ---
# Giả lập dữ liệu input từ hình ảnh (Test case k=14)
matrix = [
    [1, 2, 3, 4, 5],
    [2, 3, 4, 5, 6],
    [3, 4, 5, 45, 56]
]
k = 14

print(f"Ma trận đầu vào: {matrix}")
print(f"Tìm phần tử nhỏ thứ k={k}")

result = find_kth_smallest_binary_search(matrix, k)
print(f"Kết quả: {result}") # Output mong đợi: 45
```
