```python
def fib(n):
    if n <= 1:
        return 1
    return fib(n - 1) + fib(n - 2)

def find_fib(A, n=1):
    if fib(n) > A:
        return fib(n-1)
    return find_fib(A, n+1)

A = int(input("Nhập A: "))
print(find_fib(A))


```

```python
def solve():
    # Nhập số nguyên dương
    try:
        n_str = input("Nhập số nguyên dương: ")
        # Kiểm tra xem có phải số không
        if not n_str.isdigit():
            print("Vui lòng nhập một số hợp lệ!")
            return
            
        n = int(n_str)
        s = list(n_str) # Chuyển thành danh sách các ký tự để dễ đổi chỗ
        
        min_val = n # Giả sử số ban đầu đang là nhỏ nhất
        
        # Duyệt qua tất cả các cặp vị trí (i, j)
        for i in range(len(s)):
            for j in range(i + 1, len(s)):
                # Tạo một bản sao của danh sách để thử đổi
                temp = s.copy()
                
                # Đổi chỗ vị trí i và j
                temp[i], temp[j] = temp[j], temp[i]
                
                # Chuyển lại thành số nguyên
                # Lưu ý: int() sẽ tự xử lý trường hợp số 0 ở đầu (vd: 05 -> 5)
                new_val = int("".join(temp))
                
                # So sánh để tìm số nhỏ nhất
                if new_val < min_val:
                    min_val = new_val
        
        print(f"Số nhỏ nhất có thể đạt được là: {min_val}")

    except Exception as e:
        print("Có lỗi xảy ra:", e)

# Chạy chương trình
solve()
```

```python
def solve_sort_search():
    # Nhập dữ liệu
    n_str = input("Nhập số: ")
    original = list(n_str)           # Danh sách gốc: ví dụ ['9', '6', '2', '8']
    target = sorted(original)        # Danh sách đã xếp: ví dụ ['2', '6', '8', '9']
    
    # BƯỚC 1: TÌM KIẾM VỊ TRÍ CẦN ĐỔI
    # So sánh từng đôi một xem vị trí nào bị lệch
    idx_can_doi = -1
    val_can_tim = ''
    
    for i in range(len(original)):
        if original[i] != target[i]:
            idx_can_doi = i       # Đã tìm thấy vị trí đầu tiên chưa chuẩn (số 9)
            val_can_tim = target[i] # Giá trị đáng lẽ phải nằm ở đây (số 2)
            break # Dừng tìm kiếm ngay
            
    # Nếu idx_can_doi vẫn là -1 nghĩa là số đã sắp xếp sẵn rồi (vd: 123)
    if idx_can_doi == -1:
        print(n_str)
        return

    # BƯỚC 2: TÌM SỐ ĐÓ ĐANG NẰM Ở ĐÂU
    # Ta cần tìm số `val_can_tim` trong phần còn lại của danh sách gốc
    # Mẹo: Phải tìm từ dưới lên (từ phải sang trái) để lấy số xa nhất (nếu có trùng)
    idx_thay_the = -1
    for j in range(len(original) - 1, idx_can_doi, -1):
        if original[j] == val_can_tim:
            idx_thay_the = j
            break
            
    # BƯỚC 3: ĐỔI CHỖ (SWAP)
    original[idx_can_doi], original[idx_thay_the] = original[idx_thay_the], original[idx_can_doi]
    
    # In kết quả (dùng int để bỏ số 0 ở đầu nếu có)
    result = int("".join(original))
    print(result)

solve_sort_search()
```
