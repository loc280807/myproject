```python
#Viết hàm kiểm tra một số có phải là số nguyên tố hay không? 
import math

def ktrasnt(n):
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return n >= 2

n = int(input())
for x in range(1,n+1):
    if(ktrasnt(x)):
        print(x, end = " ")
```

```python
#VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên

def shh(n):
    sum = 0
    for i in range(1,  n+ 1):
        if n%i == 0:
            sum = sum + i
            if sum == n:
                return True

shh(6)
```

```python
#Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của
#phần tử đầu tiên có giá trị k trong list số nguyên, nếu không có thì trả về -1 


def kiemtra(A,k):
    for i in range(len(A)):
        if k == A[i]:
            return i
    return -1

A = list(map(int,input().split()))
k= int(input())
print(kiemtra(A,k))
```

```python
def pheptoan(n1, n2):
    pheptru = n1 - n2 
    phepcong = n1 + n2
    phepchia = n1 / n2
    phepnhan = n1 * n2
    return round(pheptru, 2), round(phepcong, 2), round(phepchia, 2), round(phepnhan, 2)

pheptoan(7.36, 5)
```

```python
def nhiphancua(n):
    A = []
    while n > 0:
        du = n % 2
        A.append(du)
        n = n // 2
    A.reverse()     
    return A

def chuanhoa(a,b):
    A = nhiphancua(a)
    B = nhiphancua(b)
    while len(A) != len(B): 
        if len(A) < len(B): 
            A= [0]+ A
        else:
            B=[0]+B
    return A,B

chuanhoa(13,5)

def hamming(a,b):
    A, B = chuanhoa(a, b)
    sum = 0
    for i in range(len(A)):
        if A[i] != B[i]:
            sum +=1
    return sum

hamming(48226825258,5262548488)
```
