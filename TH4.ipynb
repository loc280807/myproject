```python
# Viết chương trình nhập vào một số nguyên n và in ra màn hình số chữ số (trừ dấu)
# của số đó. Không dùng kiểu dữ liệu string
n = 0
try:
    n = int(input())
    n = abs(n)
    count = 0
    if n==0:
        count = 1
    for _ in range (10**9):
        if n > 0:
            count += 1
            n //= 10
        else:
            break
    print(count)
except ValueError:
    print("ban nhap sai du lieu")
```
